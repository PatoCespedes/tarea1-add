Ahora, considerando que todos los días no hábiles no tienen valores de dólar asignado, entonces creemos una variable que según sea el día de la semana lo calificaremos de hábil, finde (fin de semana) o feriado, aprovechando que el valor del dólar obligatoriamente va a ser NaN.

```python
def feriados(df):
    if pd.isna(df["1.Dólar observado"]) and (df["dia_de_la_semana"]=="lunes" or df["dia_de_la_semana"]=="martes" or df["dia_de_la_semana"]=="miércoles" or df["dia_de_la_semana"]=="jueves" or df["dia_de_la_semana"]=="viernes"):
        val="feriado"
    elif  pd.isna(df["1.Dólar observado"]) and (df["dia_de_la_semana"]=="sábado" or df["dia_de_la_semana"]=="domingo"):
        val="finde"
    else:
        val="hábil"
    return val
```

```python
df_bc_all_dias["feriado"]=df_bc_all_dias.apply(feriados, axis=1)
df_bc_all_dias.head(123)
```

Finalmente, crea una variable con las combinaciones [dia,mes] que tienen al menos un NaN que sea en día de semana (es decir, excluyendo Sábado y Domingo)

Calcula el promedio del precio del dolar en los días de cumpleaños del grupo para cada año entre 1983 y 2021 y el promedio del precio del dolar para cada año completo entre 1983 y 2021, y muestra una tabla donde se vean ambos resultados. ¿Qué diferencias hay en ambos indicadores en cada año?

```python
prom_anuales=df_bc_all_dias_1983_2021.groupby("year").mean()
prom_anuales.rename(columns={"1.Dólar observado":"precio_año_corrido"},inplace=True)
```

```python
birthday=["31Ago","18Ago","21Ene","16Jun","23Dic"]
print(type(birthday))
```

```python
df_bc_all_cumples=df_bc_all_dias_1983_2021[df_bc_all_dias_1983_2021["dia/mes"].isin(birthday)]
df_bc_all_cumples
```

```python
promedios_comparados = df_bc_all_cumples.groupby(['year'], as_index=False)['1.Dólar observado'].mean()
promedios_comparados.rename(columns={"1.Dólar observado":"precio_año_cumpleaños"},inplace=True)
```

```python
pd.merge(promedios_comparados,prom_anuales,on="year",how="outer")
```

Las diferencias entre el precio promedio del dolar de un mismo año corrido (todos los días hábiles de un mismo año) versus el promedio del precio en los dias del año en que los integrantes del grupo están de cumpleaños (5 días del año) son mínimas, considerando la diferencia en promedios entre un año y otro. En nuestro caso, tenemos 1 integrante de cumpleaños en enero, 1 en junio, 2 en agosto y 1 en diciembre lo que, dentro de todo, es una combinación relativamente representativa de la fluctuación anual que puede tener el precio del dolar. De ahí que las diferencias en los promedios sea relativamente pequeña.
